## Importar as bibliotecas necessárias

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Gráfico de linha 3d
Criar a figura<br>
Criar os eixos de plotagem em 3 dimensões com o parâmetros projection = '3d'<br>
Definir as coordenadas dos pontos<br>
* Para cada valor do eixo z vamos calcular as coordenadas dos eixos x e y com as funções cosseno e seno.
* Criar 1000 pontos na coordenada z variando entre 0 e 15 (muitos pontos são necesários para dar a continuidade de uma linha).
* Seno e cosseno variam entre -1 e 1. Os valores apresentados nos eixos vão de -1 a 1 (o último vlor 1.5 não é mostrado).

```python
fig = plt.figure()
ax = plt.axes(projection="3d")

z = np.linspace(0, 15, 1000)
x = np.cos(z)
y = np.sin(z)

xyticks = np.arange(-1, 1.5, 0.5)
ax.set_xticks(xyticks)
ax.set_yticks(xyticks)

ax.plot3D(x, y, z, 'blue')
plt.show()
```

## Gráfico de supeficie
### Exemplo 1 dos slides
Definir a função (os valores das coordenadas z serão calculados a partir da função):<br>
* $f_{X,Y}(x,y)=e^{-(x+y)}$<br>

Definir a grade de coordenadas dos eixos x e y:
* Em cada coordenada 30 valores entre 0 e 3
* Para cada variação da coordenada x são necessários todos os valores da coordenada y
* Usar a função meshgrid para gerar as coordenadas<br>


Definir as coordenadas do eixo z:
* As coordenadas z são calculadas com a função

```python
# Definir a função
def f(x, y):
    return np.exp(-(x+y))

# Definir a grade de coordenadas dos eixos x e Y
x = np.linspace(0, 3, 30)
y = np.linspace(0, 3, 30)
X, Y = np.meshgrid(x, y)

# Definir as coordenadas do eixo z
Z = f(X, Y)

# Criar os objetos figure e axes
fig = plt.figure(figsize=(6, 6))
ax = plt.axes(projection="3d")

# Rótulos dos eixos x, y, z
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# Título do gráfico
ax.set_title('Exemplo 1');

# Plotagem
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='winter', edgecolor='none')
plt.show()
```

### Exemplo 5 dos slides
Definir a função (os valores das coordenadas z serão calculados a partir da função):<br>
* $F_{X,Y}(x,y)=(1-e^{-x})(1-e^{-y})$<br>

Definir a grade de coordenadas dos eixos x e y:
* Em cada coordenada 30 valores entre 0 e 3
* Para cada variação da coordenada x são necessários todos os valores da coordenada y
* Usar a função meshgrid para gerar as coordenadas<br>


Definir as coordenadas do eixo z:
* As coordenadas z são calculadas com a função

```python
# Definir a função
def F(x, y):
    return (1-np.exp(-x))*(1-np.exp(-y))

# Definir a grade de coordenadas dos eixos x e y
x = np.linspace(0, 3, 30)
y = np.linspace(0, 3, 30)
X, Y = np.meshgrid(x, y)

# Definir as coordenadas do eixo z
Z = F(X, Y)

# Criar os objetos figure e axes
fig = plt.figure(figsize=(6, 6))
ax = plt.axes(projection="3d")

# Rótulos dos eixos x, y, z
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# Título do gráfico
ax.set_title('Exemplo 5');

# Plotagem
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='winter', edgecolor='none')
plt.show()
```

## Gaissiana Bivariada Padrão
Definir a função:<br>
* $f_{X,Y}(x,y)=\frac{1}{2\pi\sqrt{(1-\rho )^{2}}} \, exp\left [\frac{1}{2\sqrt{(1-\rho )^{2}}} \left ( x^{2} +2\rho xy + y^{2}\right ) \right ]$<br>

Definir a grade de coordenadas dos eixos x e y:
* Em cada coordenada 100 valores entre -4 e +4
* Para cada variação da coordenada x são necessários todos os valores da coordenada y
* Usar a função meshgrid para gerar as coordenadas<br>


Definir as coordenadas do eixo z:
* As coordenadas z são calculadas com a função

Função contour para plotar as linhas de contorno.

```python
# Definir a função
def f(x, y, ro):
    return (1/(2*np.pi*np.sqrt(1-(ro**2))))*np.exp(-(1/(2*(1-ro**2)))*(x**2-2*ro*x*y+y**2))

# Definir a grade de coordenadas dos eixos x e y
x = np.linspace(-4, 4, 100)
y = np.linspace(-4, 4, 100)
X, Y = np.meshgrid(x, y)

# Definir as coordenadas do eixo z
Z = f(X, Y, 0)

# Criar os objetos figure e axes
fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection="3d")

# Rótulos dos eixos x, y, z
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')


# Título do gráfico
ax.set_title('Gaussiana Bivariada Padrão');

ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=3, cstride=3, linewidth=1,
                cmap='winter', edgecolor='none')

#ax.set_zlim(0,0.15)

plt.show()

fig, ax = plt.subplots(figsize=(10, 10))
ax.contour(X, Y, Z)
ax.set_title('Contorno Gaussiana Bivariada Padrão')
plt.show()
```

## Gaussiana Multivariada
Função *multivariate_normal* definida em *scipy.stats*<br>
Argumentos:<br><br>
$\; \; \; \; \; \;
\mu = \begin{bmatrix}
E\left [ X_{1} \right ]\\ 
E\left [ X_{2} \right ]\\ 
\vdots \\ 
E\left [ X_{n} \right ]\end{bmatrix} \; \; \; \; \; \; \; \; \; \; \; \;
\Sigma = \begin{bmatrix}
COV\left[X_{1},X_{1}\right] & COV\left[X_{1},X_{2}\right] & \cdots  & COV\left[X_{1},X_{n}\right]\\ 
COV\left[X_{2},X_{1}\right] & COV\left[X_{2},X_{2}\right] & \cdots  & COV\left[X_{2},X_{n}\right]\\ 
\vdots  & \vdots  & \ddots  & \vdots \\ 
COV\left[X_{n},X_{1}\right] & COV\left[X_{n},X_{2}\right] & \cdots  & COV\left[X_{n},X_{n}\right]
\end{bmatrix}$<br>

Definir a grade de coordenadas dos eixos x e y:
* Em cada coordenada 100 valores entre $\mu -4\sigma $ e $\mu +4\sigma$
* Para cada variação da coordenada x são necessários todos os valores da coordenada y
* Usar a função meshgrid para gerar as coordenadas<br>


Definir as coordenadas do eixo z:
* As coordenadas z são calculadas com a função

Função contour para plotar as linhas de contorno.

```python
# Gaussiana Multivariada
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import multivariate_normal

mu1 = 0.0
mu2 = 0.0
var1 = 1.0
dp1 = np.sqrt(var1)
var2 = 1.0
dp2 = np.sqrt(var2)
cov12 = -0.9
cov21 = cov12

# Definir a função multivariada
MU = np.array([mu1, mu2])
SIGMA = np.array([[var1, cov12],[cov21, var2]])
nmv = multivariate_normal(MU, SIGMA)

# Coordenadas X, Y e Z
x = np.linspace(mu1-4*dp1, mu1+4*dp1, 100)
y = np.linspace(mu2-4*dp2, mu2+4*dp2, 100)
X, Y = np.meshgrid(x, y)
pos = np.dstack((X, Y))
Z = nmv.pdf(pos)


# Criar os objetos figure e axes
fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection="3d")

# Rótulos dos eixos x, y, z
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')


# Título do gráfico
ax.set_title('Gaussiana Multivariada');

ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=3, cstride=3, linewidth=1,
                cmap='winter', edgecolor='none')
plt.show()


fig, ax = plt.subplots(figsize=(10, 10))
ax.contour(X, Y, Z)
ax.set_title('Contorno Gaussiana Multivariada')
plt.show()
```
