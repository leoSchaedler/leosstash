### Bivariados

```python
# Criar a base
df = pd.DataFrame({
    'x' : np.random.normal(10, 1.2, 20000),
    'y' : np.random.normal(10, 1.2, 20000),
    'group' : np.repeat('A', 20000)
})

tmp1 = pd.DataFrame({
    'x' : np.random.normal(14.5, 1.2, 20000),
    'y' : np.random.normal(14.5, 1.2, 20000),
    'group' : np.repeat('B', 20000)
})

tmp2 = pd.DataFrame({
    'x' : np.random.normal(9.5, 1.5, 20000),
    'y' : np.random.normal(15.5, 1.5, 20000),
    'group' : np.repeat('C', 20000)
})

# concatenating the data
df = df.append(tmp1).append(tmp2)
```

```python
df.head(5)
```

```python
# analise com x e y
sns.scatterplot(df['x'], df['y'])
```

```python
sns.scatterplot(df['x'], df['y'], alpha=0.2, s=3)
```

```python
# densidade
sns.kdeplot(df['x'], df['y'], shade=True)
```

```python
# jointplot
sns.jointplot(df['x'], df['y'], kind='kde')
```

```python
# jointplot
sns.jointplot(df['x'], df['y'], s=2)
```

```python
# Amostragem
# pegando 1000 linhas de forma aleatoria
df_amostragem = df.sample(2000)

sns.scatterplot(df_amostragem['x'], df_amostragem['y'])
```

```python
# Ressaltar uma parte especifica dos dados
df_a = df[df['group'] == 'A']
sns.scatterplot(df['x'], df['y'], s=3)
sns.scatterplot(df_a['x'], df_a['y'], s=3, color='red')
plt.show()
```

```python
sns.scatterplot(df['x'], df['y'], hue=df['group'])
```

```python
# Facetamento
g = sns.FacetGrid(df, col='group', hue='group')
g = g.map(plt.scatter, 'x', 'y', s=2)
plt.show()
```
