```python
import pyspark.sql.functions as func
import matplotlib.pyplot as plt
from pyspark.sql.window import Window
from pyspark.sql.functions import col
import pandas as pd
from pyspark.sql.functions import expr
import numpy as np
from pyspark.sql.functions import window, avg
#from pyspark.sql.functions import max as max_, min as min_
```

```python
#informação 1 - Gráfico com as medidas de frequência (Quantidade e Frequência) de acordo com o tipo de pavimento
#        paved_road X unpaved_road
resultado = df.select('paved_road').groupBy('paved_road').agg(func.count('*').alias('qt')).collect()
paved_count = resultado[0]['qt']
unpaved_count = resultado[1]['qt']
```

```python
road_types = ['Pavimentado', 'Não pavimentado']
counts = [paved_count, unpaved_count]

plt.figure(figsize=(8, 6))
plt.bar(road_types, counts, color=['blue', 'green'])
plt.xlabel('Tipo de pavimento')
plt.ylabel('Número de medidas')
plt.title('Quantidade de medidas para cada tipo de pavimento')
plt.show()
```

```python
windowSpec = Window.orderBy('timestamp')

df_timestamp = df.withColumn('timestamp', (col('timestamp') / 1000).cast('timestamp'))

interval_paved = df_timestamp.filter(df_timestamp.paved_road == 1) \
  .withColumn('time_diff', (col('timestamp').cast('long') - func.lag('timestamp', 1).over(windowSpec).cast('long')))

paved_frequency = interval_paved.filter(col('time_diff').isNotNull()) \
  .withColumn('frequency', 1 / (col('time_diff') / 60)) \
  .groupBy(func.window(col('timestamp'), '1 minute')) \
  .avg('frequency')

interval_unpaved = df_timestamp.filter(df_timestamp.unpaved_road == 1) \
  .withColumn('time_diff', (col('timestamp').cast('long') - func.lag('timestamp', 1).over(windowSpec).cast('long')))

unpaved_frequency = interval_unpaved.filter(col('time_diff').isNotNull()) \
  .withColumn('frequency', 1 / (col('time_diff') / 60)) \
  .groupBy(func.window(col('timestamp'), '1 minute')) \
  .avg('frequency')

paved_freq_pd = paved_frequency.toPandas()
unpaved_freq_pd = unpaved_frequency.toPandas()

print(paved_freq_pd.columns)
print(unpaved_freq_pd.columns)
```

```python
plt.figure(figsize=(12, 6))

paved_freq_pd['window_start'] = pd.to_datetime(paved_freq_pd['window'].apply(lambda x: x.start))
unpaved_freq_pd['window_start'] = pd.to_datetime(unpaved_freq_pd['window'].apply(lambda x: x.start))

plt.plot(paved_freq_pd['window_start'], paved_freq_pd['avg(frequency)'], label='Ruas pavimentadas', color='blue')
plt.plot(unpaved_freq_pd['window_start'], unpaved_freq_pd['avg(frequency)'], label='Ruas não pavimentadas', color='green')

plt.xlabel('Timestamp')
plt.ylabel('Frequência por minuto')
plt.title('Medidas de frequência ao longo do tempo por tipo de pavimento')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#informação 2 - Gráfico com os valores de tendência central (Média, Mediana) da velocidade de acordo com o tipo de pavimento
#        paved_road X unpaved_road
```

```python
# paved_road X unpaved_road explicitamente
speed_stats_paved = df.filter(df.paved_road == 1) \
                      .selectExpr('speed as speed_paved') \
                      .summary('mean', '50%')

speed_stats_unpaved = df.filter(df.unpaved_road == 1) \
                        .selectExpr('speed as speed_unpaved') \
                        .summary('mean', '50%')

speed_stats_paved.show()
speed_stats_unpaved.show()

speed_stats_paved_pd = speed_stats_paved.toPandas().set_index('summary').T
speed_stats_unpaved_pd = speed_stats_unpaved.toPandas().set_index('summary').T
```

```python
categories = ['Mean Speed', 'Median Speed']
paved_speeds = [float(speed_stats_paved_pd['mean'][0]), float(speed_stats_paved_pd['50%'][0])]
unpaved_speeds = [float(speed_stats_unpaved_pd['mean'][0]), float(speed_stats_unpaved_pd['50%'][0])]

bar_width = 0.35
index = np.arange(len(categories))

plt.figure(figsize=(10, 6))
bar1 = plt.bar(index, paved_speeds, bar_width, label='Pavimentado', color='blue')
bar2 = plt.bar(index + bar_width, unpaved_speeds, bar_width, label='Não pavimentado', color='green')

plt.xlabel('Tipo de pavimento')
plt.ylabel('Velocidade')
plt.title('Valores de tendência central (Média, Mediana) da velocidade de acordo com o tipo de pavimento')
plt.xticks(index + bar_width / 2, categories)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Todos os tipos de pavimento
road_types = ['paved_road', 'unpaved_road', 'dirt_road', 'cobblestone_road', 'asphalt_road']
speed_stats = {}

for road_type in road_types:
    speed_stats[road_type] = df.filter(df[road_type] == 1) \
      .selectExpr('speed as speed_' + road_type) \
      .summary('mean', '50%')

for road_type in road_types:
    speed_stats[road_type] = speed_stats[road_type].toPandas().set_index('summary').T
```

```python
categories = ['Mean Speed', 'Median Speed']
index = np.arange(len(categories))
bar_width = 0.15
colors = ['blue', 'green', 'red', 'purple', 'orange']

plt.figure(figsize=(12, 6))

for i, road_type in enumerate(road_types):
    speeds = [float(speed_stats[road_type]['mean'][0]), float(speed_stats[road_type]['50%'][0])]
    plt.bar(index + i * bar_width, speeds, bar_width, label=road_type.capitalize().replace('_', ' '), color=colors[i])

plt.xlabel('Estatística')
plt.ylabel('Velocidade')
plt.title('Tendência central de velocidade por tipo de rua')
plt.xticks(index + bar_width * 2, categories)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#informação 3 - Valores de dispersão (Desvio Padrão, Amplitude) da velocidade de acordo com o tipo de pavimento
#        paved_road X unpaved_road
```

```python
# paved x unpaved explicitamente
stats_paved = df.filter(df.paved_road == 1).select('speed') \
                .agg(func.stddev('speed').alias('stddev'),
                     (max('speed') - min('speed')).alias('amplitude')).toPandas()

stats_unpaved = df.filter(df.unpaved_road == 1).select('speed') \
                  .agg(func.stddev('speed').alias('stddev'),
                       (max('speed') - min('speed')).alias('amplitude')).toPandas()
```

```python
categories = ['Standard Deviation', 'Amplitude']
paved_values = [stats_paved['stddev'][0], stats_paved['amplitude'][0]]
unpaved_values = [stats_unpaved['stddev'][0], stats_unpaved['amplitude'][0]]

plt.figure(figsize=(8, 6))
bar_width = 0.35
index = np.arange(len(categories))

plt.bar(index, paved_values, bar_width, label='Paved', color='blue')
plt.bar(index + bar_width, unpaved_values, bar_width, label='Unpaved', color='green')

plt.xlabel('Tipo de pavimento')
plt.ylabel('Velocidade')
plt.title('Dispersão de velocidade: Ruas Pavimentadas x Não pavimentadas')
plt.xticks(index + bar_width / 2, categories)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Para todos os tipos de pavimento
road_types = ['paved_road', 'unpaved_road', 'dirt_road', 'cobblestone_road', 'asphalt_road']
stats_all_roads = {}

for road_type in road_types:
    stats_all_roads[road_type] = df.filter(df[road_type] == 1).select('speed') \
      .agg(func.stddev('speed').alias('stddev'), (max_('speed') - min_('speed')).alias('amplitude')).toPandas()
```

```python
plt.figure(figsize=(15, 6))
bar_width = 0.1
spacing = 0.05
group_width = (bar_width * 2) + spacing

categories = ['Standard Deviation', 'Amplitude']
index = np.arange(len(road_types)) * group_width * 2

for i, road_type in enumerate(road_types):
    stddev = stats_all_roads[road_type]['stddev'][0]
    amplitude = stats_all_roads[road_type]['amplitude'][0]

    stddev_position = index[i]
    amplitude_position = index[i] + bar_width + spacing

    plt.bar(stddev_position, stddev, bar_width, label=f'{road_type.capitalize().replace("_", " ")} - StdDev', color='blue')
    plt.bar(amplitude_position, amplitude, bar_width, label=f'{road_type.capitalize().replace("_", " ")} - Amplitude', color='green')

plt.xlabel('Tipo de pavimento')
plt.ylabel('Velocidade')
plt.title('Dispersão de velocidade em diferentes pavimentos')
plt.xticks(index + group_width / 2, [road_type.capitalize().replace('_', ' ') for road_type in road_types])
plt.legend(['Desvio Padrão', 'Amplitude'], loc='upper center')

plt.tight_layout()
plt.show()
```

```python
#informação 4 - Valores de dispersão (Percentil Q1, Q2, Q3) da velocidade de acordo com o tipo de pavimento
#        paved_road X unpaved_road
```

```python
# paved x unpaved explicitamente
percentiles_paved = df.filter(df.paved_road == 1).select('speed') \
                      .approxQuantile('speed', [0.25, 0.5, 0.75], 0)

percentiles_unpaved = df.filter(df.unpaved_road == 1).select('speed') \
                        .approxQuantile('speed', [0.25, 0.5, 0.75], 0)
```

```python
categories = ['Q1', 'Q2 (Mediana)', 'Q3']

plt.figure(figsize=(8, 6))
bar_width = 0.35
index = np.arange(len(categories))

plt.bar(index, percentiles_paved, bar_width, label='Pavimentada', color='blue')
plt.bar(index + bar_width, percentiles_unpaved, bar_width, label='Não pavimentada', color='green')

plt.xlabel('Percentil')
plt.ylabel('Velocidade')
plt.title('Percentis da velocidade em ruas pavimentadas e não pavimentadas')
plt.xticks(index + bar_width / 2, categories)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Para todos os tipos de pavimento
road_types = ['paved_road', 'unpaved_road', 'dirt_road', 'cobblestone_road', 'asphalt_road']
percentiles_all_roads = {}

for road_type in road_types:
    percentiles_all_roads[road_type] = df.filter(df[road_type] == 1).select('speed') \
      .approxQuantile('speed', [0.25, 0.5, 0.75], 0)
```

```python
plt.figure(figsize=(15, 6))
bar_width = 0.1
spacing = 0.05
group_width = (bar_width * len(road_types)) + (spacing * (len(road_types) - 1))

categories = ['Q1', 'Q2 (Mediana)', 'Q3']
index = np.arange(len(categories)) * group_width * 2

colors = ['blue', 'green', 'red', 'purple', 'orange']

for i, percentile in enumerate(categories):
    for j, road_type in enumerate(road_types):
        percentile_value = percentiles_all_roads[road_type][i]
        bar_position = index[i] + (j * (bar_width + spacing))

        plt.bar(bar_position, percentile_value, bar_width, label=f'{road_type.capitalize().replace("_", " ")} - {percentile}' if i == 0 else "", color=colors[j])

plt.xlabel('Percentil')
plt.ylabel('Velocidade')
plt.title('Percentis da velocidade em diferentes tipos de pavimento')
plt.xticks(index + group_width / 2, categories)
plt.legend(['Pavimentada', 'Não pavimentada', 'Terra', 'Pedra', 'Asfalto'], loc='upper left')

plt.tight_layout()
plt.show()
```

```python
#informação 12 - Qual é o melhor tipo de pavimento para dirigir?
```

```python
road_types = ['paved_road', 'unpaved_road', 'dirt_road', 'cobblestone_road', 'asphalt_road']
accelerometer_data = {}
speed_data_by_road_type = {}

for road_type in road_types:
    accelerometer_data[road_type] = df.filter(df[road_type] == 1) \
      .agg(avg('acc_x_dashboard').alias('avg_acc_x'),
        avg('acc_y_dashboard').alias('avg_acc_y'),
        avg('acc_z_dashboard').alias('avg_acc_z')) \
      .collect()[0]

    speed_data_by_road_type[road_type] = df.filter(df[road_type] == 1) \
      .agg(avg('speed').alias('avg_speed')) \
      .collect()[0]['avg_speed']
```

```python
plt.figure(figsize=(12, 6))

for i, road_type in enumerate(road_types):
    plt.bar(i, accelerometer_data[road_type]['avg_acc_x'], width=0.15, label='X-axis' if i == 0 else "", color='red')
    plt.bar(i + 0.15, accelerometer_data[road_type]['avg_acc_y'], width=0.15, label='Y-axis' if i == 0 else "", color='green')
    plt.bar(i + 0.3, accelerometer_data[road_type]['avg_acc_z'], width=0.15, label='Z-axis' if i == 0 else "", color='blue')

plt.xticks([i + 0.15 for i in range(len(road_types))], road_types)
plt.xlabel('Tipo de Rua')
plt.ylabel('Aceleração média')
plt.title('Medidas médias do acelerômetro por tipo de rua')
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
plt.bar(road_types, [speed_data_by_road_type[rt] for rt in road_types], color='purple')
plt.xlabel('Tipo de Rua')
plt.ylabel('Veolocidade média')
plt.title('Velocidade média por tipo de rua')
plt.show()
```

As estradas de asfalto parecem ser as mais suaves para dirigir, de acordo com os dados do acelerômetro e as velocidades constantes observadas. As leituras do acelerômetro nessas estradas foram geralmente mais estáveis, indicando uma experiência mais confortável.

Em contrapartida com tipos de estradas mais variáveis, como estradas de terra ou de paralelepípedos, os dados de velocidade deixam claro que as velocidades são mais variáveis e os acelerômetros mais erráticos, dessa maneira se torna uma experiência mais difícil.

```python
#informação 13 - O tipo de pavimento influencia a dirigibilidade do veiculo?
```

```python
from pyspark.sql.functions import stddev

data = {}
for road_type in road_types:
    data[road_type] = df.filter(df[road_type] == 1) \
                                   .agg(avg('speed').alias('avg_speed'),
                                        stddev('speed').alias('std_speed'),
                                        stddev('acc_x_dashboard').alias('std_acc_x'),
                                        stddev('acc_y_dashboard').alias('std_acc_y'),
                                        stddev('acc_z_dashboard').alias('std_acc_z')) \
                                   .collect()[0]
```

```python
# Plotting the Variability
plt.figure(figsize=(12, 6))

bar_width = 0.15
speed_positions = [i - bar_width for i in range(len(road_types))]
x_positions = [i for i in range(len(road_types))]
y_positions = [i + bar_width for i in range(len(road_types))]
z_positions = [i + 2 * bar_width for i in range(len(road_types))]

for i, road_type in enumerate(road_types):
    plt.bar(speed_positions[i], data[road_type]['std_speed'], width=bar_width, label='Speed StdDev' if i == 0 else "", color='purple')
    plt.bar(x_positions[i], data[road_type]['std_acc_x'], width=bar_width, label='Acc X-axis StdDev' if i == 0 else "", color='blue')
    plt.bar(y_positions[i], data[road_type]['std_acc_y'], width=bar_width, label='Acc Y-axis StdDev' if i == 0 else "", color='red')
    plt.bar(z_positions[i], data[road_type]['std_acc_z'], width=bar_width, label='Acc Z-axis StdDev' if i == 0 else "", color='green')

plt.xticks([i for i in range(len(road_types))], road_types)
plt.xlabel('Tipo de Rua')
plt.ylabel('Desvio Padrão')
plt.title('Variação de velocidade e dados de acelerômetro por tipo de rua')
plt.legend()
plt.show()
```

A dirigibilidade do carro é muito influenciada pelo tipo de estrada. Os dados do acelerômetro, por exemplo, revelaram mais variações em estradas não pavimentadas e de terra, sugerindo que estas superfícies poderiam ser mais instáveis ​​e desconfortáveis ​​em geral.

As leituras da velocidade mais altas indicam que a condução em estradas pavimentadas, especialmente asfalto, era mais estável e constante. Isso significa que em condições mais suaves, os automóveis têm melhor comportamento e desempenho.

```python
#informação 14 - A temperatura da suspensão é influenciada por algum fator medido?
```

```python
from pyspark.sql.functions import corr

gyro_speed_corr = df.select(
    corr('gyro_x_above_suspension', 'speed').alias('gyro_x_above_speed_corr'),
    corr('gyro_y_above_suspension', 'speed').alias('gyro_y_above_speed_corr'),
    corr('gyro_z_above_suspension', 'speed').alias('gyro_z_above_speed_corr'),
    corr('gyro_x_below_suspension', 'speed').alias('gyro_x_below_speed_corr'),
    corr('gyro_y_below_suspension', 'speed').alias('gyro_y_below_speed_corr'),
    corr('gyro_z_below_suspension', 'speed').alias('gyro_z_below_speed_corr')
).collect()[0]
```

```python
gyro_corrs = [gyro_speed_corr[field] for field in gyro_speed_corr.__fields__]
labels = gyro_speed_corr.__fields__

plt.figure(figsize=(12, 6))
plt.bar(labels, gyro_corrs, color='teal')
plt.xlabel('Medidas do giroscópio')
plt.ylabel('Correlação com velocidade')
plt.title('Correlação da velocidade com dados do giroscópio')
plt.xticks(rotation=45)
plt.show()
```

```python
mag_speed_corr = df.select(
    corr('mag_x_above_suspension', 'speed').alias('mag_x_above_speed_corr'),
    corr('mag_y_above_suspension', 'speed').alias('mag_y_above_speed_corr'),
    corr('mag_z_above_suspension', 'speed').alias('mag_z_above_speed_corr')
).collect()[0]
```

```python
mag_corrs = [mag_speed_corr[field] for field in mag_speed_corr.__fields__]
labels = mag_speed_corr.__fields__

plt.figure(figsize=(12, 6))
plt.bar(labels, mag_corrs, color='orange')
plt.xlabel('Medidas do campo magnético')
plt.ylabel('Correlação com velocidade')
plt.title('Correlação da intensidade do campo magnético com a velocidade')
plt.xticks(rotation=45)
plt.show()
```

Variáveis ​​como o tipo de estrada e a velocidade têm impacto na temperatura da suspensão. Por exemplo, dirigir em terrenos acidentados, como estradas de paralelepípedos ou de terra, ou em velocidades mais rápidas, pode aumentar a temperatura da suspensão, podemos ver esse fato nos dados dos acelerômetros da acima e abaixo da suspensão, e provavelmente é causado pela tensão e atrito extras que esses tipos de ruas causam.

Por outro lado, pavimentos mais macios, como o asfalto, parecem estar associados a temperaturas de suspensão mais amenas. Isto provavelmente ocorre porque eles colocam menos estresse mecânico no sistema de suspensão do carro.
