```python
class Person:
  def __init__(self, nome, idade):
    self.nome = nome
    self.idade = idade
  def __str__(self):
    return f'Nome: {self.nome} \t Idade: {self.idade}'
```

```python
l = []
l.append(Person('Jean', 28))
l.append(Person('Jose Eduardo', 18))
l.append(Person('Pavao', 17))
```

```python
#incrementar a idade de todas as pessoas 
l = list(map(lambda x: Person(x.nome, x.idade + 1), l))
for obj in l:
  print(obj)
```

```python
# selecionar pessoas com mais de 20 anos
l_menor_20 = list(filter(lambda x: x.idade> 20, l))
for obj in l_menor_20:
  print(obj)
```

```python
#somar todas as idades
soma = 0
for obj in l:
  soma += obj.idade
soma
```

```python
from functools import reduce

reduce(lambda x, y: x + y, map(lambda x: x.idade, l))
```

# Pair RDDs - Contagem de palavras

```python
# carregamento de arquivo

rdd = sc.textFile('file://' + SparkFiles.get('bible.txt'))
rdd.take(5)
```

Implementação anterior:


*   Carregava o rdd
*   Cada item do rdd era uma frase
*   flatMap -> cada frase em n palabvras
*   Contagem das palavras: countByValue



Implementação atual:


*   Carregar o rdd
*   flatMap -> cada frase em palavras
*   Gerar para cada palavra (palavra, 1)
*   Somar as ocorrências de cada palavra
*   A lógica segue o MapReduce do Hadoop



```python
#quebrando as linahs em palavras
palavras = rdd.flatMap(lambda x: x.split(" "))
```

```python
#cirando (palavra, 1)
prdd = palavras.map(lambda x: (x, 1))
```

```python
prdd.take(5)
```

```python
#somar as ocorrências
contagem = prdd.reduceByKey(lambda x, y: x + y)
```

```python
contagem.take(10)
```

```python
contagem_invertida = contagem.map(lambda t: (t[1], t[0]))
```

```python
ordenado = contagem_invertida.sortByKey(ascending=False)
```

```python
contagem_ordenada = ordenado.map(lambda t: (t[1], t[0]))
```

```python
contagem_ordenada.take(50)
```

# Real estate

```python
#carregamento de arquivo
rdd = sc.textFile('file://' + SparkFiles.get('RealEstate.csv'))
```

```python
# ignorando o cabecalho
rdd = rdd.filter(lambda x: not x.startswith('MLS,Location,'))
```

```python
# indices
IX_PRECO = 2
IX_QUARTOS = 3

def monta_tupla(s):
  vals = s.split(',')
  # gerando tupla no formato (número de quartos, (preço da casa, quantidade de casa = 1))
  return (int(vals[IX_QUARTOS]), (float(vals[IX_PRECO]), 1))
```

```python
# criando um pair rdd usando a função que gera tuplas
pair_rdd = rdd.map(monta_tupla)
```

```python
pair_rdd.take(3)
```

```python
# somar os preços e as ocorrências
# x e y são os valores compostos (preço, ocorrência)
# para somar x e y, precisamos somar seus componentes individualmente
# (preço + preço) e (ocorrência + ocorrência)
somas = pair_rdd.reduceByKey(lambda x, y: (x[0] + y[0], x[1] + y[1]))
```

```python
somas.take(5)
```

```python
# dividindo soma dos preços pela quantidade de casas
medias = somas.mapValues(lambda t: t[0] / t[1])
medias.collect()
```

```python
# ordenando os resultados
ordenado = medias.sortByKey(ascending=True)
ordenado.collect()
```

```python
agrupado = pair_rdd.groupByKey()
agrupado.mapValues(list).collect()
```
